```python
#creating a numpy array
my_list=[1,2,3,4,5,6]
import numpy as np
array=np.array(my_list,dtype=int)
print(array)
```

```python
print(len(array)) # length of array
print(type(array))# tupe of the array
print(array.ndim)# diamenstions of the array
print(array.shape)# shape of array
arr=array.reshape(3,2) #reshaping the array
print(arr)
print(arr.shape)
```
